```python
from keras import Sequential
from keras.layers import Dense,SimpleRNN
```

```python
model = Sequential()
model.add(SimpleRNN(3,input_shape=(4,5))) # 4 is the no of timestap,s and 5 is the no of input features
model.add(Dense(1,activation='sigmoid'))
model.summary()
```

```python
print(model.get_weights()[0].shape)
model.get_weights()[0]
```
